```python
col_hig_14_019 = 'blue'
col_hig_14_022 = 'green'
col_atlas = 'red'
```

```python

plt.savefig('exclusion_sigma_BR.pdf')
```

```python
plt.plot(df['m_a [GeV]'], df['xsec * BR / xsec (SM) (HIG-14-019)'], 
         label='CMS 4'+r'$\tau$'+' HIG-14-019', color=col_hig_14_019)
# stupid hack cos it doesn't like the first entry being NaN
mask = df.isnull()['xsec * BR / xsec (SM) (HIG-14-022)']
plt.plot(df[~mask]['m_a [GeV]'], df[~mask]['xsec * BR / xsec (SM) (HIG-14-022)'], 
         label='CMS 4'+r'$\tau$'+' HIG-14-022', color=col_hig_14_022)
plt.plot(df['m_a [GeV]'], df['xsec * BR / xsec (SM) (ATLAS)'], 
         label='ATLAS '+r'$\mu\mu\tau\tau$', color=col_atlas)
plt.yscale('log')

ax = plt.gca()

# shade the regions
upper_edge_14_019 = np.ones_like(df['xsec * BR / xsec (SM) (HIG-14-019)'])*ax.get_ylim()[1]
ax.fill_between(df['m_a [GeV]'], df['xsec * BR / xsec (SM) (HIG-14-019)'], 
                y2=upper_edge_14_019, 
                where=df['xsec * BR / xsec (SM) (HIG-14-019)']<upper_edge_14_019,
                label='CMS HIG-14-019 limit', alpha=0.4, color=col_hig_14_019)

upper_edge_14_022 = np.ones_like(df[~mask]['xsec * BR / xsec (SM) (HIG-14-022)'])*ax.get_ylim()[1]
ax.fill_between(df[~mask]['m_a [GeV]'], df[~mask]['xsec * BR / xsec (SM) (HIG-14-022)'], 
                y2=upper_edge_14_022, 
                where=df[~mask]['xsec * BR / xsec (SM) (HIG-14-022)']<upper_edge_14_022,
                label='CMS HIG-14-022 limit', alpha=0.2, color=col_hig_14_022)

upper_edge_ATLAS = np.ones_like(df['xsec * BR / xsec (SM) (ATLAS)'])*ax.get_ylim()[1]
ax.fill_between(df['m_a [GeV]'], df['xsec * BR / xsec (SM) (ATLAS)'], 
                y2=upper_edge_ATLAS, 
                where=df['xsec * BR / xsec (SM) (ATLAS)']<upper_edge_ATLAS,
                label='ATLAS limit', alpha=0.2, color=col_atlas)


plt.xlabel(r'$m_a\ [\mathrm{GeV}]$')
plt.ylabel(r'$\frac{\sigma}{\sigma_{SM}} \times BR(pp \to h_{125} \to 2a \to 4\tau)$')

plt.ylim(top=2, bottom=0.02)
plt.legend(loc=4)
plt.minorticks_on()
plt.savefig('exclusion_BR.pdf')
```
